```python
import csv

users_cohort = {}
start_cohort = 160
end_cohort = 167

for i in range(start_cohort,end_cohort+1):
    with open("/home/gtrindadi/jds_june_2019_copy/data-sendatree/folder2/day_"+str(i), "r") as f:
        for row in csv.reader(f,delimiter=' '):
            # se o usuário (row[2]) não está na list de cohort: adicione ele com a data de envio (row[1])
            if not(row[2] in users_cohort):
                users_cohort[row[2]] = [row[1]]
            # se o usuário (row[2]) está na lista de cohort, mas ainda não tem aquela data de envio (row[1]): adicione a ele a data de envio
            elif (row[2] in users_cohort) and not(row[1] in users_cohort[row[2]]):
                users_cohort[row[2]].append(row[1])
            else:
                continue
    f.close()
```

```python
print("user;","free_tree_sent_date")
for keys,values in users_cohort.items():
    print(keys+": "+str(values))

#print(users_cohort)
```

```python
import csv

with open('users-cohort-set-output.csv', 'w') as csvFile:
    writer = csv.writer(csvFile)
    writer.writerow(['user','sent_dates'])
    
    for key in users_cohort.keys():
        writer.writerow([key] + users_cohort[key])
csvFile.close()
```
