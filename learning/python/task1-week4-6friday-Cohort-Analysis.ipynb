```python
# reading files and putting the date organized in a dictionary
import csv

users_cohort = {}
start_cohort = 1
end_cohort = 30

for i in range(start_cohort,end_cohort+1):
    cohort = "day_"+str(i)
    with open("/home/gtrindadi/jds_june_2019/data-sendatree/data-free-tree/"+cohort, "r") as f:
        for row in csv.reader(f,delimiter=' '):
            # se o usuário (row[2]) não está na list de cohort: adicione ele com a data de envio (row[1] or row[0])
            if not(row[2] in users_cohort):
                users_cohort[row[2]] = [row[0]]
            # se o usuário (row[2]) está na lista de cohort, mas ainda não tem aquela data de envio (row[1] or row[0]): adicione a ele a data de envio
            elif (row[2] in users_cohort) and not(row[0] in users_cohort[row[2]]):
                users_cohort[row[2]].append(row[0])
            else:
                continue
    f.close()
```

```python
# the cohort analysis
day_list = []
result = [["cohort,days_passed,number_of_users"]]

for i in range(start_cohort,end_cohort+1):
    day_list.append('day_'+str(i))

for cohort in day_list:
    day_passed = 0
    # deve-se olhar da data do cohort em diante, pois queremos saber de quem enviou a partir do cohort até os próximos 7 dias
    for given_day in day_list[day_list.index(cohort):day_list.index(cohort)+8]:
        number_of_users = 0
        for user_dates in users_cohort.values():
            if (given_day in user_dates) and (user_dates[0] == cohort):
                number_of_users += 1
        result.append([cohort + ',' + str(day_passed) + ',' + str(number_of_users)])
        day_passed = day_passed + 1
        
result
```

```python
# putting the cohort analysis in a file
with open('cohort-analysis-output.csv', 'w') as csvFile:
    writer = csv.writer(csvFile)
    
    for i in result:
        writer.writerow(i)

csvFile.close()
```
